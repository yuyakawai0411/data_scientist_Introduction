```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(0.0, 1.0, 10**4)
y = np.random.uniform(0.0, 1.0, 10**4)

plt.figure(figsize=(20, 6))
plt.subplot(2, 1, 1)
plt.plot(x)

plt.subplot(2, 1, 2)
plt.plot(y)
```

```python
x.size
```

```python
inside_x = []
inside_y = []
outside_x = []
outside_y = []

count_inside = 0
for count in range(10**4):
    d = math.hypot(x[count], y[count])
    if d < 1:
        count_inside += 1
        inside_x.append(x[count])
        inside_y.append(y[count])
    else:
        outside_x.append(x[count])
        outside_y.append(y[count])
print("", count_inside)
```

```python
plt.figure(figsize=(6,6))
plt.plot(inside_x, inside_y, 'o')
plt.plot(outside_x, outside_y, 'o')
plt.grid(True)
```
