```python
import matplotlib as mpl
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
%matplotlib inline
```

```python
np.random.seed(0)
x = np.random.randn(30)
y = np.sin(x) + np.random.randn(30)
plt.figure(figsize=(6,4))
# グラフの作画
plt.plot(x, y, 'o') # 'o'は点をプロットするという意味
plt.title('Title Name')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
```

```python
np.random.seed(0)
x = np.arange(1000)
y = np.random.randn(1000).cumsum()
plt.figure(figsize=(6,4))
# # グラフの作画
plt.plot(x, y, label='Label')
plt.legend()
plt.title('Title Name')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
```

```python
plt.figure(figsize=(20,6))
# 2行一列のグラフの1つ目
plt.subplot(2,1,1)
x = np.linspace(-10, 10, 100) # -10から10を100分割したリストを表示する
plt.plot(x, np.sin(x))

# 2行一列のグラフの2つ目
plt.subplot(2,1,2)
y = np.linspace(-10, 10, 100) # -10から10を100分割したリストを表示する
plt.plot(y, np.sin(2*y))

plt.grid(True)
```

**二次関数グラフの作図**

```python
def my_function(x):
    return x**2 + 2*x + 1

x = np.arange(-10, 10)
plt.figure(figsize=(20,6))
plt.plot(x, my_function(x))
plt.grid(True)
```

```python
np.random.seed(0)
plt.figure(figsize=(20, 6))
y = np.random.randn(10 ** 5) * 10 + 50
# binsはヒストグラムの棒を区切る単位。大きいほど細く、小さいほど太くなる
plt.hist(y, bins=60, range=(20, 80))
plt.grid(True)
```

## 練習問題

```python
def my_function(x):
    return 5*x + 3
x = np.linspace(-10, 10, 100)
plt.figure(figsize=(20,6))
plt.plot(x, my_function(x))
plt.grid(True)
```

```python
x = np.linspace(-10, 10, 100)
plt.figure(figsize=(6, 4))
# subplotを使わなければplotした数だけ、グラフが重なる
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.grid(True)
```

```python
def uni_hist(N):
    x = np.random.uniform(0.0, 1.0, N)
    y = np.random.uniform(0.0, 1.0, N)

    plt.subplot(2, 1, 1)
    plt.hist(x)
    plt.title('No1:histgram')
    plt.grid(True)

    plt.subplot(2, 1, 2)
    plt.hist(y)
    plt.title('No2:histgram')
    plt.grid(True)
    plt.tight_layout()

uni_hist(1000)
```
