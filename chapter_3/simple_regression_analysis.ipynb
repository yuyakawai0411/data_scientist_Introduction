```python
import pandas as pd
import matplotlib.pyplot as plt
student_data_math = pd.read_csv('sample_data/student-mat.csv', sep=';')
```

```python
plt.plot(student_data_math['G1'], student_data_math['G3'], 'o')
plt.xlabel('G1 grid')
plt.ylabel('G3 grid')
plt.grid(True)
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
```

```python
X = student_data_math.loc[:, ['G1']].values
Y = student_data_math['G3'].values
# fitの第一引数には、2次元データを渡さなけれあならない => Expected 2D array, got 1D array instead:
reg.fit(X, Y)
```

```python
reg.coef_
```

```python
plt.plot(X, Y, 'o')
plt.xlabel('G1 grid')
plt.ylabel('G3 grid')

plt.plot(X, reg.predict(X))
plt.grid(True)
```

```python
# 決定係数
reg.score(X, Y)
```

```python
student_data_por = pd.read_csv('sample_data/student-por.csv', sep=';')
```

## 練習問題

```python
plt.plot(student_data_por['G1'], student_data_por['G3'], 'o')
plt.xlabel('G1 grid')
plt.ylabel('G3 grid')
plt.grid(True)
```

```python
X = student_data_por.loc[:, ['G1']].values
Y = student_data_por['G3'].values
reg = linear_model.LinearRegression()
reg.fit(X, Y)

print('回帰係数', reg.coef_)
print('切片', reg.intercept_)
print('決定係数', reg.score(X, Y))
```

```python
X = student_data_por.loc[:, ['absences']].values
Y = student_data_por['G3'].values
reg = linear_model.LinearRegression()
reg.fit(X, Y)

print('回帰係数', reg.coef_)
print('切片', reg.intercept_)
print('決定係数', reg.score(X, Y))

plt.plot(student_data_por['absences'], student_data_por['G3'], 'o')
plt.xlabel('absences')
plt.ylabel('G3')

plt.plot(X, reg.predict(X))
plt.grid(True)
```
